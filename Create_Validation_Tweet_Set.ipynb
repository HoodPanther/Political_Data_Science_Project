```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
tweet300=pd.ExcelFile('300 Tweet Sample.xlsx')
tweet300_df=tweet300.parse('Sheet2')
tweet300_df=tweet300_df.drop(['Candidate'],axis=1)
```

```python
tweet500_1=pd.read_csv('Clinton Tweets 2.csv')
tweet500_1_df=tweet500_1.iloc[:1000,:]
tweet500_2=pd.read_csv('Trump Tweets 2.csv')
tweet500_2_df=tweet500_2.iloc[:1000,:]
```

```python
combined_df=pd.concat([tweet300_df,tweet500_1_df,tweet500_2_df],join='outer',axis=0).iloc[:,[5,4,2,1,3,6,0]]
combined_df=combined_df[pd.isnull(combined_df.iloc[:,6])]
```

```python
combined_df=combined_df.drop(['Different Code (by Prof. Leblang)'],axis=1)
combined_df=combined_df.fillna(0)
```

```python
combined_df['Neutral']=combined_df.iloc[:,3]+combined_df.iloc[:,4]+combined_df.iloc[:,5]
```

```python
combined_df_2=combined_df.drop(['Mixed','Neither/Neutral',"unclear/can't code"],axis=1)
```

```python
category={(1,0,0):'Positive',(0,1,0):'Negative',(0,0,1):'Neutral',(0,0,0):'Error'}
```

```python
sentiment=['']*combined_df_2.shape[0]
for i in range(combined_df_2.shape[0]):
    sentiment[i]=category[tuple(combined_df_2.iloc[i,1:])]
```

```python
combined_df_2['sentiment']=sentiment
```

```python
combined_df_2=combined_df_2[combined_df_2.sentiment!='Error']
```

```python
combined_df_2.drop(['Positive','Negative','Neutral'],axis=1).to_csv('sentiment_validation.csv',header=True,index=False)
```
