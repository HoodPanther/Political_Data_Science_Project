### (1) Candidate Follower Counts

```python
cf_counts_list=glob.glob("Candidate Follower Counts/Raw/2015*")+glob.glob("Candidate Follower Counts/Raw/2016*")
```

```python
remove_list=[]
cf_counts_list=sorted(cf_counts_list)
for i,cf in enumerate(cf_counts_list):
    try:
        if re.search("2016\-06|2016\-07|2016\-08|2016\-05\-31",cf).group(0)!='':
            remove_list.append(i)
    except AttributeError:
        pass
remove_list=np.array(remove_list)
```

```python
cf_counts_list=np.array(cf_counts_list)
whole_list=np.array(list(range(len(cf_counts_list))))
cf_counts_list=cf_counts_list[np.setdiff1d(whole_list,remove_list)]
```

```python
all_cf_tables=[None]*len(cf_counts_list)
for i,cf in enumerate(cf_counts_list):
    all_cf_tables[i]=pd.read_csv(cf)
```

```python
combined_table=pd.concat(all_cf_tables,axis=0,join="outer")
combined_table.tail(30)
```

```python
combined_table=combined_table.iloc[np.setdiff1d(np.array(range(combined_table.shape[0])),np.where(pd.isnull(combined_table.Follower_count))[0]),:]
```

```python
combined_table.Follower_count=combined_table.Follower_count.astype(int)
combined_table
```

```python
candidate_list=list(set(combined_table.Candidate))
for c in candidate_list:
    single_candidate=combined_table[combined_table.Candidate==c].iloc[:,[0,2,3,4]]
    single_candidate.to_csv("Candidate Follower Counts/Aggregated/"+c+" - Candidate Follower Counts.csv",index=False)
```

### (2) Celebrity Endorser Follower Counts

```python
cef_counts_list=glob.glob("Celebrity Endorser Follower Counts/Raw/2015*")+glob.glob("Celebrity Endorser Follower Counts/Raw/2016*")+glob.glob("Celebrity Endorser Follower Counts/Raw/Wikidemo*")
```

```python
remove_list=[]
cef_counts_list=sorted(cef_counts_list)
for i,cef in enumerate(cef_counts_list):
    try:
        if re.search("2016\-08|Wikidemo2016\-08",cef).group(0)!='':
            remove_list.append(i)
    except AttributeError:
        pass
remove_list=np.array(remove_list)
```

```python
cef_counts_list=np.array(cef_counts_list)
whole_list=np.array(list(range(len(cef_counts_list))))
cef_counts_list=cef_counts_list[np.setdiff1d(whole_list,remove_list)]
```

```python
all_cef_tables=[None]*len(cef_counts_list)
for i,cef in enumerate(cef_counts_list):
    all_cef_tables[i]=pd.read_csv(cef,encoding="ISO-8859-1") #needs to reset the encoding here
```

```python
def replace_invalid_value(x):
    try:
        if re.search("Error in",str(x)).group(0)!="":
            return '0'
    except AttributeError:
        return str(x)
    
def replace_null_value(x):
    if pd.isnull(x):
        return 0
    else:
        return x
```

```python
combined_table=pd.concat(all_cef_tables,axis=0,join="outer").sort_values(by=["Candidates","Date"])
combined_table=combined_table[~pd.isnull(combined_table['Follower_count'])]
combined_table['Follower_count']=combined_table['Follower_count'].apply(replace_invalid_value).apply(replace_null_value).apply(int)
combined_table=\
combined_table.groupby(["Candidates","Date"])['Follower_count'].agg(['count','sum','max','mean','median']).reset_index()\
.rename(columns={"sum":"total_follower_counts","max":"maximum_follower_counts","mean":"avg_follower_counts",\
                "median":"median_follower_counts"}).iloc[:,[0,1,3,4,5,6]]
combined_table['avg_follower_counts']=combined_table['avg_follower_counts'].astype(int)
combined_table['median_follower_counts']=combined_table['median_follower_counts'].astype(int)
combined_table
```

```python
candidate_list=list(set(combined_table.Candidates))
for c in candidate_list:
    combined_table[combined_table.Candidates==c].iloc[:,1:].to_csv("Celebrity Endorser Follower Counts/Aggregated/"+c+" - Celebrity Endorser Follower Counts.csv",index=False)
```

### (3) Political Endorser Follower Counts

```python
pef_counts_list=glob.glob("Political Endorser Follower Counts/Raw/538demo*")+glob.glob("Political Endorser Follower Counts/Raw/R538count*")
```

```python
remove_list=[]
pef_counts_list=sorted(pef_counts_list)
for i,pef in enumerate(pef_counts_list):
    try:
        if re.search("2016\-08",pef).group(0)!='':
            remove_list.append(i)
    except AttributeError:
        pass
remove_list=np.array(remove_list)
```

```python
pef_counts_list=np.array(pef_counts_list)
whole_list=np.array(list(range(len(pef_counts_list))))
pef_counts_list=pef_counts_list[np.setdiff1d(whole_list,remove_list)]
```

```python
all_pef_tables=[None]*len(pef_counts_list)
for i,pef in enumerate(pef_counts_list):
    all_pef_tables[i]=pd.read_csv(pef,encoding="ISO-8859-1") #needs to reset the encoding here
    if 'date' in all_pef_tables[i].columns.values:
        all_pef_tables[i] = all_pef_tables[i].drop('Date', 1)
        all_pef_tables[i]=all_pef_tables[i].rename(columns={"date":"Date"})
```

```python
combined_table=pd.concat(all_pef_tables,axis=0,join="outer").sort_values(by=["Candidates","Date"])
combined_table=combined_table[~pd.isnull(combined_table['Follower_count'])]
for i in range(combined_table.shape[0]):
    if pd.isnull(combined_table.iloc[i,1]):
        combined_table.iloc[i,1]=combined_table.iloc[i,0]
```

```python
combined_table_2=combined_table.iloc[:,1:]
combined_table_2['Follower_count']=combined_table_2['Follower_count'].apply(replace_invalid_value).astype(float).astype(int)
combined_table_2=\
combined_table_2.groupby(["Candidates","Date"])['Follower_count'].agg(['count','sum','max','mean','median']).reset_index()\
.rename(columns={"sum":"total_follower_counts","max":"maximum_follower_counts","mean":"avg_follower_counts",\
                "median":"median_follower_counts","count":"number_of_endorsers"})
combined_table_2['avg_follower_counts']=combined_table_2['avg_follower_counts'].astype(int)
combined_table_2['median_follower_counts']=combined_table_2['median_follower_counts'].astype(int)
combined_table_2.ix[np.where(combined_table_2["Date"]=="10/22/15")[0],"Date"]='2015-10-22'
combined_table_2=combined_table_2.sort_values(by=["Candidates","Date"])
combined_table_2
```

```python
candidate_list=list(set(combined_table_2.Candidates))
for c in candidate_list:
    combined_table_2[combined_table_2.Candidates==c].iloc[:,1:].to_csv("Political Endorser Follower Counts/Aggregated/"+c+" - Political Endorser Follower Counts.csv",index=False)
```
