```python
import numpy as np
import pandas as pd
import glob
import re
import os
os.chdir("/home/composersyf/Documents/Political Data Science Project/OtherData")
```

```python
Democratic_files_list=glob.glob("PredictWise/Raw/Predictwise-Democratic*.csv")
Republican_files_list=glob.glob("PredictWise/Raw/Predictwise-Republic*.csv")
Democratic_files_list.sort()
Republican_files_list.sort()
```

```python
d_remove_list=[]
Democratic_files_list=sorted(Democratic_files_list)
for i,f in enumerate(Democratic_files_list):
    try:
        if re.search("2016\-08",f).group(0)!='':
            d_remove_list.append(i)
    except AttributeError:
        pass
d_remove_list=np.array(d_remove_list)
print(d_remove_list)

r_remove_list=[]
Republican_files_list=sorted(Republican_files_list)
for i,f in enumerate(Republican_files_list):
    try:
        if re.search("2016\-08",f).group(0)!='':
            r_remove_list.append(i)
    except AttributeError:
        pass
r_remove_list=np.array(r_remove_list)
print(r_remove_list)
```

```python
Democratic_files_list=np.array(Democratic_files_list)
d_whole_list=np.array(list(range(len(Democratic_files_list))))
Democratic_files_list=Democratic_files_list[np.setdiff1d(d_whole_list,d_remove_list)]

Republican_files_list=np.array(Republican_files_list)
r_whole_list=np.array(list(range(len(Republican_files_list))))
Republican_files_list=Republican_files_list[np.setdiff1d(r_whole_list,r_remove_list)]
```

```python
None_array=np.array([None]*15)
None_array=None_array.reshape(1,15)

Democratic_Data=pd.DataFrame(None_array)
Democratic_Data.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",
             "PredictIt_Buy","PredictIt_Sell","Bookie_Odds","Hypermind","Hypermind_Bid","Hypermind_Ask","date"]
Democratic_Data
```

```python
for file_name in Democratic_files_list:
    file=pd.read_csv(file_name)
    file=file.iloc[:,1:]
    if file.shape[1]==14:
        file.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",\
                     "PredictIt_Buy","PredictIt_Sell","Bookie_Odds","Hypermind","Hypermind_Bid","Hypermind_Ask"]
    elif file.shape[1]==10:
        file.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",\
                     "PredictIt_Buy","PredictIt_Sell"]
        file["Bookie_Odds"]=["Not existing"]*file.shape[0]
        file["Hypermind"]=["Not existing"]*file.shape[0]
        file["Hypermind_Bid"]=["Not existing"]*file.shape[0]
        file["Hypermind_Ask"]=["Not existing"]*file.shape[0]
    elif file.shape[1]==11:
        file.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",\
                     "PredictIt_Buy","PredictIt_Sell","Bookie_Odds"]
        file["Hypermind"]=["Not existing"]*file.shape[0]
        file["Hypermind_Bid"]=["Not existing"]*file.shape[0]
        file["Hypermind_Ask"]=["Not existing"]*file.shape[0]
    date=file_name.split(".")[0][::-1][:10][::-1]
    file['date']=[date]*file.shape[0]
    Democratic_Data=pd.concat([Democratic_Data,file],join="outer")
Democratic_Data=Democratic_Data.iloc[1:,]
Democratic_Data=Democratic_Data.sort_values(["date","Candidate"])
Democratic_Data.head(20)
```

```python
Democratic_candidates=Democratic_Data['Candidate'].astype(np.ndarray).unique()
for c in Democratic_candidates:
    Democratic_Data[Democratic_Data.Candidate==c].to_csv("PredictWise/Aggregated/"+c+" - PredictWise.csv", index=False)
```
