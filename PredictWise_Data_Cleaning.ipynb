```python
import numpy as np
import pandas as pd

import os
os.chdir("/home/composersyf/Documents/Political Data Science Project/Predictwise")
```

```python
import glob
Democratic_files_list=glob.glob("*Democratic*.csv")
Republican_files_list=glob.glob("*Republic*.csv")
Democratic_files_list.sort()
Republican_files_list.sort()
```

```python
None_array=np.array([None]*15)
None_array=None_array.reshape(1,15)

Democratic_Data=pd.DataFrame(None_array)
Democratic_Data.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",
             "PredictIt_Buy","PredictIt_Sell","Bookie_Odds","Hypermind","Hypermind_Bid","Hypermind_Ask","date"]
Democratic_Data
```

```python
for file_name in Democratic_files_list:
    file=pd.read_csv(file_name)
    file=file.iloc[:,1:]
    file.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",
                 "PredictIt_Buy","PredictIt_Sell","Bookie_Odds","Hypermind","Hypermind_Bid","Hypermind_Ask"]
    date=file_name.split(".")[0][::-1][:10][::-1]
    file['date']=[date]*file.shape[0]
    Democratic_Data=pd.concat([Democratic_Data,file])
Democratic_Data=Democratic_Data.iloc[1:,]
Democratic_Data.head(20)
```

```python
Republican_Data=pd.DataFrame(None_array)
Republican_Data.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",
             "PredictIt_Buy","PredictIt_Sell","Hypermind","Hypermind_Bid","Hypermind_Ask","Bookie_Odds","date"]
Republican_Data
```

```python
for file_name in Republican_files_list:
    file=pd.read_csv(file_name)
    file=file.iloc[:,1:]
    file.columns=['Candidate',"PredictWise","Betfair","PredictIt","Bookie","Pollster","Betfair_Back","Betfair_Lay",
                 "PredictIt_Buy","PredictIt_Sell","Hypermind","Hypermind_Bid","Hypermind_Ask","Bookie_Odds"]
    date=file_name.split(".")[0][::-1][:10][::-1]
    file['date']=[date]*file.shape[0]
    Republican_Data=pd.concat([Republican_Data,file])
Republican_Data=Republican_Data.iloc[1:,]
Republican_Data['Hypermind_Bid']=Republican_Data['Hypermind_Bid'].map('$ {:}'.format)
Republican_Data['Hypermind_Ask']=Republican_Data['Hypermind_Ask'].map('$ {:}'.format)
Republican_Data=Republican_Data.iloc[:,list(range(10))+[13]+[10,11,12,14]]
Republican_Data.head(20)
```

```python
Republican_candidates=Republican_Data['Candidate'].astype(np.ndarray).unique()
for candidate_name in Republican_candidates:
    Republican_Data[Republican_Data.Candidate==candidate_name].to_csv("Republican/"+candidate_name+".csv")
```

```python
Democratic_candidates=Democratic_Data['Candidate'].astype(np.ndarray).unique()
for candidate_name in Democratic_candidates:
    Democratic_Data[Democratic_Data.Candidate==candidate_name].to_csv("Democratic/"+candidate_name+".csv")
```
